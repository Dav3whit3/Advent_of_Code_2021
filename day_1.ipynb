```python
import numpy as np
from functools import reduce
import requests
```

```python
s = requests.Session()
session_id = "53616c7465645f5fa3d5749a329093b03d094af1c181e8b5837f44f167feaa71d424a72db2beda1f3b85147dda382bb2"
r = s.get("https://adventofcode.com/2021/day/1/input", cookies={'session': session_id})
res = r.text.split("\n")
depths = list(int(x) for x in res if x != '')
```

# PART 1

```python
len(depths)
```

```python
from functools import reduce
reduce(lambda a, b: a+1 if b>0 else a+0, np.diff(depths), 0)
```

# PART 2

```python
counter = 0
track = 0
for index, elem in enumerate(depths):
    if index == 0:
        track = reduce(lambda a, b: a+b, depths[0:3])
    else:
        if len(depths) - index >= 3:
            g = reduce(lambda a, b: a+b, depths[index:index+3])
            if g > track:
                counter += 1
            track = g
print(f"{counter=}")
```
